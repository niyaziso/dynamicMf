Visualize the value of objective function in dynamicMF for each iteration with different values of hidden dimensions K (K = 3, 5 and 10)

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the reconstruction errors from log file

# K = 3
errors3 = []
with open('./Data/log/error_3.txt') as f:
    for l in f:
        errors3.append(float(l.strip()))
# K = 5
errors5 = []
with open('./Data/log/error_5.txt') as f:
    for l in f:
        errors5.append(float(l.strip()))
# K = 10
errors10 = []
with open('./Data/log/error_10.txt') as f:
    for l in f:
        errors10.append(float(l.strip()))
```

```python
mx = np.array([errors3[-1],errors5[-1],errors10[-1]]) # get errors in the last iteration
dn = 1000*1709*86 # number of entries in tensor
mx = np.sqrt(mx/dn) # compute RMSE
```

```python
# make the plot
fig, ax = plt.subplots()
ax.plot(errors3,'--')
ax.plot(errors5,'-.')
ax.plot(errors10)
ax.legend(['K=3','K=5','K=10'],loc='upper center')
ax.set_xlabel('Iterations')
ax.set_ylabel('Objective Function Value')
left, bottom, width, height = [0.7, 0.65, 0.2, 0.2]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.bar([1,2,3],mx)
ax2.set_xticks([1,2,3])
ax2.set_xticklabels([3,5,10])
plt.tight_layout()
fig.savefig('optimization_error.eps',dpi=192)
```
